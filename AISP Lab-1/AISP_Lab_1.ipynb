```python
import librosa
from IPython.display import Audio, display
filename = "AI in speech processing.m4a"
y, sr = librosa.load(filename)
librosa.display.waveshow(y)
```

```python
# 1st Segment
filename = "AI in speech processing.m4a"
y1, sr1 = librosa.load(filename, offset=1.0, duration=1.0)
audio_segment1 = Audio(y1, rate=sr1)
display(audio_segment1)
```

```python
# 2nd Segment
y2, sr2 = librosa.load(filename, offset=1.75, duration=3)
audio_segment2 = Audio(y2, rate=sr2)
display(audio_segment2)
```
